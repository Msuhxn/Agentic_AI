# Agentic Content Workflow Demo

This notebook supports the Bachelor thesis prototype **Agentic AI for Automated Content Planning and Generation: A Lightweight Evaluation of Opportunities and Limitations**.

## Thesis Context

The prototype compares a conventional one-prompt baseline with a simple agentic workflow. The agentic workflow separates the task into visible stages such as planning, drafting, reviewing, and refining. The purpose is to support a qualitative bachelor-level comparison, not to build a production system.
